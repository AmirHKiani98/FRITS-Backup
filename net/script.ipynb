```python
import pandas as pd
import networkx as nx

# Load the CSV (adjust the path as needed)
df = pd.read_csv("plain.edg.csv", sep=";")  # Youâ€™re using semicolons, not commas

# Create a directed graph
G = nx.DiGraph()

# Add edges from CSV
for _, row in df.iterrows():
    print(row["edge_from"])
    from_node = str(row['edge_from'])  # Cast to string if needed
    to_node = str(row['edge_to'])
    G.add_edge(from_node, to_node)
```

```python
G.out_edges("0")
```

```python
nx.draw_networkx(G)
```

```python
import sumolib
import matplotlib.pyplot as plt
net = sumolib.net.readNet('4x4.net.xml')

pos = {}
for node in net.getNodes():
    pos[node.getID()] = node.getCoord()  # (x, y)
nx.draw(G, pos=pos, with_labels=True, node_size=300, node_color='lightgreen', arrows=True)
plt.show()
```

```python
for junction in net.getNodes():
    print(f"Junction {junction.getID()} connects to:")
    for edge in junction.getOutgoing():
        print(f"  â†’ {edge.getID()}")
```
